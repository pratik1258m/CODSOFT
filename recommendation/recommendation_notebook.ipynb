## 📖 Introduction to Recommendation Systems
A **Recommendation System** is a subclass of information filtering systems that seeks to predict the "rating" or "preference" a user would give to an item. They are critical tools for companies like Netflix, Amazon, and YouTube to drive user engagement.

### Recommendation Techniques
1. **Content-Based Filtering**: Suggests items similar to those a user liked in the past. It relies entirely on item properties/metadata (e.g., genre, description, keywords). 
2. **Collaborative Filtering**: Suggests items based on the preferences of similar users. It relies entirely on historical user-item interactions (e.g., ratings, purchases) and operates on the assumption that "if User A and User B agree on an issue, they are likely to agree on others."

### Mathematical Formulations
- **TF-IDF (Term Frequency-Inverse Document Frequency)**: A statistical measure evaluating how relevant a word is to a document in a collection.
  $$\text{TF-IDF}(t, d, D) = \text{TF}(t, d) \times \log\left(\frac{|D|}{1 + |\{d \in D : t \in d\}|}\right)$$
- **Cosine Similarity**: Measures the cosine of the angle between two multi-dimensional feature vectors, indicating direction similarity regardless of vector magnitude.
  $$\text{Similarity}(A, B) = \cos(\theta) = \frac{A \cdot B}{\|A\| \|B\|} = \frac{\sum_{i=1}^n A_i B_i}{\sqrt{\sum_{i=1}^n A_i^2} \sqrt{\sum_{i=1}^n B_i^2}}$$

## 🛠️ Step 1: Content-Based Filtering Implementation

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create sample movie dataset
movies_data = [
    {"id": 1, "title": "The Dark Knight", "genres": "Action Thriller Crime", "description": "Batman battles Joker in Gotham."},
    {"id": 2, "title": "Inception", "genres": "Action Sci-Fi Thriller", "description": "Thief steals secrets through dream sharing."},
    {"id": 3, "title": "Interstellar", "genres": "Sci-Fi Drama Adventure", "description": "Explorers travel through a wormhole in space."},
    {"id": 4, "title": "The Matrix", "genres": "Action Sci-Fi", "description": "Hacker Neo discovers cyber underworld reality."},
    {"id": 5, "title": "The Notebook", "genres": "Romance Drama", "description": "Passionate young lovers separated by class differences."}
]
df = pd.DataFrame(movies_data)

# Combine genres and descriptions
df['metadata'] = df['genres'] + " " + df['description']

# Calculate TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['metadata'])

# Calculate Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim, index=df['title'], columns=df['title'])

print("Cosine Similarity Matrix:")
cosine_sim_df
```

## 👥 Step 2: Collaborative Filtering Implementation

```python
import numpy as np

# Create dense User-Item rating matrix (ratings 1 to 5)
ratings_data = {
    "The Dark Knight": [5, 4, 1, 5],
    "Inception":       [5, 5, 2, 4],
    "Interstellar":    [4, 5, 1, 3],
    "The Matrix":      [5, 4, 2, 5],
    "The Notebook":    [1, 1, 4, 2]
}
ratings_df = pd.DataFrame(ratings_data, index=["User A", "User B", "User C", "User D"])

# Normalize ratings around average user rating (row-wise subtraction)
normalized_ratings = ratings_df.sub(ratings_df.mean(axis=1), axis=0)

# Compute User-to-User similarity df
user_sim = cosine_similarity(normalized_ratings)
user_sim_df = pd.DataFrame(user_sim, index=ratings_df.index, columns=ratings_df.index)

print("User-to-User Correlation Matrix:")
user_sim_df
```

## 🎓 Conclusion
- **Content-Based Filtering** is highly effective for recommending highly specific niche items because it doesn't require user logs (cold start solution for items).
- **Collaborative Filtering** captures complex behavioral patterns and cross-genre interests that description tags can never capture, but suffers from the **Cold Start Problem** (cannot recommend new items with zero ratings).
