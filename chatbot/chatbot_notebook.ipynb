## 📖 Introduction to Rule-Based Chatbots
A **rule-based chatbot** is a simple conversational agent that operates on pre-defined paths and rules. It uses text processing techniques—such as pattern matching, regular expressions, and string analysis—to parse user queries, determine intent, and select the best response from a set of pre-written answers.

### How It Works
1. **Input Normalization:** Clean and standardize the user's text (e.g., lowercasing, removing extra whitespaces, stripping punctuation).
2. **Pattern Matching:** Scan the input using **Regular Expressions (Regex)** to find predefined keywords or phrase patterns.
3. **Intent Resolution:** Map the matched pattern to a specific conversation intent (e.g., standard greeting, telling a joke, asking for name).
4. **Response Generation:** Select a response (often randomized among a list of variations to sound more natural) and format it (e.g., inserting captured groups like user names).
5. **Fallback:** If no patterns are matched, return a polite default answer to guide the user back to supported topics.

## 🛠️ Step 1: Import Required Libraries
We will use Python's built-in `re` module for regular expressions, `random` for picking dynamic responses, and `datetime` to answer queries about the current time.

```python
import re
import random
import datetime
```

## ⚙️ Step 2: Define Rules and Patterns
We create a structured set of patterns and responses. We'll define a class `RuleBasedChatbot` that compiles these patterns at initialization for fast searching.

```python
class RuleBasedChatbot:
    def __init__(self):
        # The rules list contains tuples of (regex_pattern, list_of_responses)
        self.rules = [
            (
                r"\b(hi|hello|hey|greetings|hola)\b",
                [
                    "Hello! I am your friendly rule-based assistant. How can I help you today?",
                    "Hi there! How's your day going? What can I do for you?",
                    "Hey! Nice to meet you. How can I assist you today?"
                ]
            ),
            (
                r"\b(good)?bye|see\s+ya|farewell|exit|quit\b",
                [
                    "Goodbye! Have a wonderful day ahead!",
                    "Bye! Feel free to chat again whenever you'd like.",
                    "Farewell! Take care and talk to you soon!"
                ]
            ),
            (
                r"what is your name|who are you",
                [
                    "I am Nexus Bot, a rule-based AI built for the CODSOFT AI Internship!",
                    "You can call me Nexus Bot. I'm a helper chatbot running on predefined rules."
                ]
            ),
            (
                r"how are you|how is it going|how\'s it going|how do you do",
                [
                    "I'm doing great, thank you for asking! I'm ready to answer your questions.",
                    "I'm operating at peak efficiency! How are you doing today?",
                    "Fantastic! Just waiting to help out. How about yourself?"
                ]
            ),
            (
                r"what can you do|help|features|commands",
                [
                    "I can chat with you, tell jokes, give the current time, and answer basic questions! Try asking: 'tell me a joke', 'what is the time', or 'who made you'."
                ]
            ),
            (
                r"who (created|made|built) you|source code",
                [
                    "I was created by Pratik, an AI engineering intern at CODSOFT!",
                    "My logic was built by Pratik using Python and regular expression pattern matching."
                ]
            ),
            (
                r"tell me a joke|joke|funny",
                [
                    "Why do programmers wear glasses? Because they can't C#!",
                    "There are 10 types of people in the world: those who understand binary, and those who don't.",
                    "Why did the computer go to the doctor? Because it had a virus!",
                    "How many programmers does it take to change a light bulb? None, that's a hardware problem."
                ]
            ),
            (
                r"time|date|clock",
                [
                    f"The current local time is {datetime.datetime.now().strftime('%I:%M %p')}.",
                    f"My clock says it's {datetime.datetime.now().strftime('%H:%M:%S')} right now."
                ]
            ),
            (
                r"\b(love|like) you\b",
                [
                    "Aww, thank you! I enjoy chatting with you too.",
                    "That's so kind of you! I'm here to help you as much as I can."
                ]
            ),
            (
                r"weather|rain|sunny|temp",
                [
                    "I don't have real-time internet access to check the weather, but I hope it's beautiful and sunny where you are!",
                    "I can't check the weather sensor right now, but a perfect indoor temperature for coding is 21°C!"
                ]
            ),
            (
                r"my name is ([a-zA-Z\s]+)",
                [
                    "Nice to meet you, {0}! How can I help you today?",
                    "Hello {0}! That's a wonderful name. What's on your mind today?"
                ]
            ),
            (
                r"do you know about ([a-zA-Z\s0-9]+)",
                [
                    "Yes, {0} is a fascinating topic! While I only have predefined rules, you can learn a lot about {0} by studying standard documentation or asking a search engine.",
                    "I've heard about {0}! As a rule-based bot, my knowledge is limited, but {0} sounds highly interesting."
                ]
            )
        ]
        
        # Compile patterns with re.IGNORECASE for performance and case-insensitivity
        self.compiled_rules = [(re.compile(p, re.IGNORECASE), resp) for p, resp in self.rules]
        
        # Fallback responses
        self.fallbacks = [
            "I'm not sure I understand that. Can you try rephrasing? (Type 'help' to see what I can do!)",
            "Interesting query! However, my rule-based brain doesn't have a match for that. Ask me a joke instead!",
            "I'm a simple rule-based chatbot. Could you ask something else, like 'what is your name' or 'tell me a joke'?",
            "Hmm, that's beyond my current rules. Try typing 'help' to see my capabilities!"
        ]

    def respond(self, user_input):
        user_input_clean = user_input.strip()
        if not user_input_clean:
            return "Please say something! I'm listening."
            
        # Iterate sequentially to find the first matching rule
        for pattern, responses in self.compiled_rules:
            match = pattern.search(user_input_clean)
            if match:
                response = random.choice(responses)
                # Dynamic formatting if capture groups are present
                if match.groups():
                    group_val = match.group(1).strip()
                    return response.format(group_val)
                return response
                
        # Fallback
        return random.choice(self.fallbacks)
```

## 🧪 Step 3: Test the Chatbot with Various Inputs
Let's instantiate our chatbot and test it on a series of common queries to see how it responds.

```python
# Instantiate chatbot
bot = RuleBasedChatbot()

# Predefined test queries representing different intents
test_queries = [
    "Hello there!",
    "Who are you?",
    "My name is Alex",
    "How is it going?",
    "What is the current time?",
    "Can you tell me a funny joke?",
    "Do you know about machine learning?",
    "What is the weather outside?",
    "This is a query that shouldn't match anything.",
    "See you later!"
]

# Run and print results
for query in test_queries:
    response = bot.respond(query)
    print(f"User: {query}")
    print(f"Bot:  {response}")
    print("-" * 50)
```

## 🎓 Conclusion
Rule-based chatbots are a powerful tool for **highly structured domains** (e.g., FAQ routing, basic command triggers). 

**Advantages**:
- 100% predictable; no risk of "hallucinations" or generating inappropriate/offensive language.
- Extremely lightweight; requires no GPU, no training, and executes in microseconds.
- Simple to debug and maintain.

**Limitations**:
- Cannot handle variation in language style outside the defined rules (fails at semantic understanding).
- High maintenance as the rule set scales (risk of conflicts between rules).
- Lacks memory or state tracking across long conversation turns (unless explicitly coded in a state machine).
